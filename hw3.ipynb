```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression, Ridge, Lasso
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
```

# 방의 수와 집값

```python
boston = load_boston()
```

```python
X=boston.data
y=boston.target
```

```python
x=[i[5] for i in X]
```

```python
reg=LinearRegression()

reg.fit(X,y)
```

```python
plt.plot(x,y,'o',label='Data')
plt.plot(x,reg.predict(X),'o',label='Predicted')
plt.legend()
```

```python
A=[[i] for i in x]
B=[[i] for i in reg.predict(X)]
```

```python
poly=PolynomialFeatures(degree=1,include_bias=False)
X_poly=poly.fit_transform(A)
```

```python
reg2=LinearRegression()
reg2.fit(X_poly,B)
```

```python
plt.plot(x,y,'o',label='Data')
plt.plot(x,reg2.predict(X_poly),label='Trendline')
plt.legend();
```

집값과 방의 개수의 데이터를 reg.predict로 분석해 앞으로 형성될 집값과 방의 개수의 추세 그래프를 만들었다. 그 결과 앞으로도 원래 데이터셋과 유사하게 꽤나 선형 모양의 추세를 보이며 집값이 형성될 것을 예상할 수 있다.
